```python
tags_list = insta_tags[0].split(' ')
```

```python
stop_words = ['소통','daily','데일리','좋아요','맞팔','좋반','좋아요반사','selfie',
              '셀피','ootd','오오티디','셀카','선팔','팔로우','fff','f4f','follow',
             '선팔하면 맞팔','선팔맞팔','selca','맞팔','like','선팔하면맞팔','맞팔',
             '맞팔해요','소통','선팔','좋아요테러','오오티디']

tokens = [word for word in tags_list if word not in stop_words]

ko = nltk.Text(tokens, name='국내여행 태그')
ko.vocab().most_common(200)
```

```python
def make_colors(word, font_size, position, orientation, random_state, **kwargs):
    r = random_state.randint(0,255)
    g = random_state.randint(0,255)
    b = random_state.randint(0,255)
    color = 'rgb(%d, %d, %d)'%(r,g,b)
    return color

data = ko.vocab().most_common(500)
wordcloud = WordCloud(font_path = 'c:/Windows/Fonts/malgun.ttf',
                     relative_scaling=0.5,
                     background_color='white',
                      max_font_size=50,
                     ).generate_from_frequencies(dict(data))
# gen = wordcloud.generate(data)
recolor = wordcloud.recolor(color_func = make_colors, random_state=True)
plt.figure(figsize=(20,8))
plt.imshow(recolor, interpolation='bilinear')
plt.axis('off')
plt.savefig('wordcloud_domestic_trip_insta_tags')
```
